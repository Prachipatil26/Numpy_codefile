##Handling missing values 

Builtin Function 

np.isnan= detect missing values.

np.nan_to_num() = replace null valued with number.

np.isinf() = detect infinite values.

NaN = not a number.



```python
#np.isnan(array)
import numpy as np

arr = np.array([1,2,np.nan , 4 , np.nan , 6])
print(np.isnan(arr))
print(np.isnan(arr).sum())# counting the missing value
```

```python
#np.nan_to_num() = replace null valued with number.
import numpy as np
arr = np.array([1,2,np.nan , 4 , np.nan , 6])
cleaned_arr=np.nan_to_num(arr ,nan=100)
print(cleaned_arr)
```

```python
#default value is 0
arr = np.array([1,2,np.nan , 4 , np.nan , 6])
cleaned_arr=np.nan_to_num(arr)
print(cleaned_arr)
```

```python
#detect infinite number #np.isinf()
arr=np.array([1,2,np.inf , 4 , -np.inf , 6])
print(np.isinf(arr))
```

```python
#replace infinte values
arr = np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr).sum())
cleaned_arr = np.nan_to_num(arr , posinf=1000 , neginf=1000)
print(cleaned_arr)
```
